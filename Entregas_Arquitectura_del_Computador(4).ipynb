**Algoritmo**
De acuerdo con la solicitud relacionada en la actividad de la semana 3 se debe generar un algoritmo que dé solución al problema: “Determinar si un número es primo o no”, para está actividad he decidido crear el algoritmo en el lenguaje de programación Java.


public static void main(String[] args) throws IOException {
    Main.numeroPrimo();
}

public static void numeroPrimo() throws IOException {
    BufferedReader reader = new BufferedReader(new InputStreamReader(System.in));
    System.out.println("Ingresar un número");
    int entrada = Integer.parseInt(reader.readLine());
    int primo = 0;
    for(int i = 1; i <= entrada; i++) {
        if (entrada % i == 0) {
            primo++;
        }
    }
    if (primo == 2) {
        System.out.println("El número " + entrada + " es primo");
    }else {
        System.out.println("El número " + entrada + " no es primo");
    }
}

**Comprobación de ejemplos probados y resultados**

Luego de completar la creación del programa en logisim, obtenemos un ALU que contiene la lógica para detectar cuando un número es primo o no es primo al obtener un resultado luego de hacer varias operaciones aritmeticas, para determinar que operación se utilizará en cada proceso lo haremos a tráves de un multiplexor.

* **A** = Primer número binario con el que se operara
* **B** = Segundo número binario con el que se operara
* **Código de Operación** = Binario con el que determinará la operación (**Suma = 000** / **Resta = 001** / **División = 010**)
* **Res_Binario** = Binario con el que se mostrará el resultado de la operación
* **Número primo** = Valor con el que se identificara cuando es un milagro (Si es primo = 1 / Si no es primo = 0)

* **Primer Número (A)** = Primer Número en entero
* **Primer Número (B)** = Segundo Número en entero
* **Resultado** = Es el valor que se obtendra como resultado de las diferentes operaciones (Valor mostrado en entero)
